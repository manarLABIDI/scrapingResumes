```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url1 = 'https://www.timesjobs.com'

location = ''
page = 1
joblist = []
# Define the search parameters

post_weeks = 60
search_type = 'personalizedSearch'
result_size = 25
search_by = 0
operator = 'OR'
query = 'linux'

posting_date = 'I'

for sequence in range(1, 39):
    if sequence % 10 == 1:
        page = 1
    else:
        page = ((sequence // 10) * 10) + 1
        print (page)

    # Build the URL for the current page
    url = f'https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords={query}&searchBy={search_by}&rdoOperator={operator}&searchType={search_type}&luceneResultSize={result_size}&postWeek={post_weeks}&txtKeywords={query}&pDate={posting_date}&sequence={sequence}&startPage={page}'
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    divs = soup.find_all('li', {'class': 'clearfix job-bx wht-shd-bx'})
    print(len(divs))
    for pt in divs:
            link = pt.find('a')['href']
            
            #print(link)
            r2 = requests.get(link)
            soup2 = BeautifulSoup(r2.content, 'html.parser')
            title_elem = soup2.find('h1', {'class': 'jd-job-title'})
            title = title_elem.text.strip() if title_elem else 'N/A'
            
            print(title)
            
            
            required_skills = pt.find('span', {'class': 'srp-skills'}).text.strip()
            #print(required_skills)

            experience = pt.find('ul', {'class': 'top-jd-dtl'}).find('li').text.replace('card_travel', '')
            #print(experience)

            ul_tag = soup2.find('ul', {'class': 'clearfix', 'id': 'applyFlowHideDetails_1'})
            li_tags = ul_tag.find_all('li', {'class': 'clearfix'})
            li_3 = li_tags[3]
            li = li_3.find('span', {'class': 'basic-info-dtl'})
            if li is not None:
            
                graduation = li.text.strip()
                #print(graduation)
            else:
                graduation = 'Any Graduate'
            job = {
                'title'    : title,
                'required_skills' : required_skills,
                'experience': experience,
            
                'graduation'   : graduation
            }  
            #print(job)
            joblist.append(job) 
            print(len(joblist))
            df = pd.DataFrame(joblist)

            #print(df.head())
            df.to_csv('jobs_7.csv')


```
